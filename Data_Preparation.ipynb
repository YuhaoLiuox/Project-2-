```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as ss
import tensorflow as tf
import math
import random
```

```python
year=[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
```

```python
stand=pd.read_excel('df2009.xlsx')
stan=np.unique(stand['industryName1'])
```

```python
df1=[]

for y in year:
    d1=pd.read_excel('df'+str(y)+'.xlsx')
    del d1['Unnamed: 0']
    for i in range(len(stan)):
        for j in range(len(d1['industryName1'])):
            if d1['industryName1'][j]==stan[i]:
                d1.loc[j,'isNew']=i
                
    d1['date']=y*100+1
    d1['year']=y
    d1['mon']=1
    df1.append(d1)
```

```python
df1[0]
```

```python
df2=[]

for y in year:
    d1=pd.read_excel('df'+str(y)+'q2'+'.xlsx')
    del d1['Unnamed: 0']
    for i in range(len(stan)):
        for j in range(len(d1['industryName1'])):
            if d1['industryName1'][j]==stan[i]:
                d1.loc[j,'isNew']=i
    d1['date']=y*100+2
    d1['year']=y
    d1['mon']=2    
    df2.append(d1)
```

```python
df2[0]
```

```python
df3=[]

for y in year:
    d1=pd.read_excel('df'+str(y)+'q3'+'.xlsx')
    del d1['Unnamed: 0']
    for i in range(len(stan)):
        for j in range(len(d1['industryName1'])):
            if d1['industryName1'][j]==stan[i]:
                d1.loc[j,'isNew']=i
    d1['date']=y*100+3
    d1['year']=y
    d1['mon']=3
    df3.append(d1)
```

```python
df3[0]
```

```python
df4=[]

for y in year:
    d1=pd.read_excel('df'+str(y)+'q4'+'.xlsx')
    del d1['Unnamed: 0']
    for i in range(len(stan)):
        for j in range(len(d1['industryName1'])):
            if d1['industryName1'][j]==stan[i]:
                d1.loc[j,'isNew']=i
    d1['date']=y*100+4
    d1['year']=y
    d1['mon']=4
    df4.append(d1)
```

```python
df4[0]
```

```python
df_con1=pd.concat(df1).reset_index()
del df_con1['index']
df_con1
```

```python
df_con2=pd.concat(df2).reset_index()
del df_con2['index']
df_con2
```

```python
df_con3=pd.concat(df3).reset_index()
del df_con3['index']
df_con3
```

```python
df_con4=pd.concat(df4).reset_index()
del df_con4['index']
df_con4
```

```python
data=pd.concat([df_con1,df_con2,df_con3,df_con4])
```

```python
data.to_excel('data.xlsx')
```
